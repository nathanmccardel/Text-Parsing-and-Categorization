# Function for Categorizing Errors

```python
def CategorizeCell(x):
    listkeywordcount = []
    for i in listoflists:
        keywordcount = 0
        for keyword in i:
            if keyword in x:
                keywordcount += 1
        listkeywordcount.append(keywordcount)
    maxnum = 0
    index = 1000
    for j in range(0,len(listkeywordcount)):
        if listkeywordcount[j] > maxnum:
            maxnum = listkeywordcount[j]
            index = j
    if index == 1000:
        categorization = 'This error cannot be categorized'
    else:
        for item in dictforlists:
            if dictforlists[item] == listoflists[index]:
                categorization = item
               # categorization = 'This is an error with ' + item
    return categorization
```

### Test Function

```python
CategorizeCell('2 for 4 dollar promotion')
```

## Function with Alternative Prioritization

```python
def CategorizePriority(x):
    matchfound = 0
    for i in range(0,len(listoflists)):
        for keyword in listoflists[i]:
            if keyword in x:
                matchfound = 1
                for item in dictforlists:
                    if dictforlists[item] == listoflists[i]:
                        categorization = item
                        return categorization
    if matchfound == 0:
        categorization = 'This error cannot be categorized'
        return categorization
```

## Create Combined Function

```python
def CategorizeCombined(x):
    matchfound = 0
    for i in range(0,len(prioritylistoflists)):
        for keyword in prioritylistoflists[i]:
            if keyword in x:
                matchfound = 1
                for item in prioritydictforlists:
                    if prioritydictforlists[item] == prioritylistoflists[i]:
                        categorization = item
                        return categorization
    if matchfound == 0:
        listkeywordcount = []
        for i in restlistoflists:
            keywordcount = 0
            for keyword in i:
                if keyword in x:
                    keywordcount += 1
            listkeywordcount.append(keywordcount)
        maxnum = 0
        index = 1000
        for j in range(0,len(listkeywordcount)):
            if listkeywordcount[j] > maxnum:
                maxnum = listkeywordcount[j]
                index = j
        if index == 1000:
            categorization = 'This error cannot be categorized'
        else:
            for item in restdictforlists:
                if restdictforlists[item] == restlistoflists[index]:
                    categorization = item
                   # categorization = 'This is an error with ' + item
        return categorization
```

## Now to try and combine phrases

```python
commonphrases = {}

def PhraseCount(x):
    if x[0] == ' ':
        x = x[1:]
    if x[len(x)-1] == ' ':
        x = x[0:len(x)-1]
    foundphrase =  0
    for key,value in commonphrases.items():
        if x == key:
            foundphrase = 1
            commonphrases[x] = commonphrases[x] + 1
    if foundphrase != 1:
        commonphrases[x] = 1
```
